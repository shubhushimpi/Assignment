```python
class Standard:
    def __init__(self):
        self.__std_name=''
        self.__no_of_enrolled=""
        self.__overallGPA=""
    def setStandard(self):
        self.__std_name=input("Enter Student Name: ")
        self.__no_of_enrolled=int(input("Enter No of Enrolled: "))
        self.__overallGPA=float(input("Enter Overall GPA: "))

    def showStandard(self):
        print("Std Name: ",self.__std_name)
        print("No of enrolled: ",self.__no_of_enrolled)
        print("Overall GPA: ",self.__overallGPA)

class Student(Standard):
    def __init__(self):
        self.setStandard()
        self.__rollno=''
        self.__math =''
        self.__english = ''
        self.__science = ''
        self.__social = ''
        
    def setStudent(self):
        self.showStandard()
        self.__rollno=int(input("Enter Roll Number : "))
        self.__math =int(input("Enter Math Marks : "))
        self.__english = int(input("Enter English Marks : "))
        self.__science = int(input("Enter Science Marks : "))
        self.__social = int(input("Enter Social Marks : "))
        
    def showStudent(self):
        print("Enter Roll No: ",self.__rollno)
        print("Enter Math Marks: ",self.__math)
        print("Enter English Marks: ",self.__english)
        print("Enter Science Marks : ",self.__science)
        print("Enter Social Marks: ",self.__social)
    
    def printResult(self):
        self.setStudent()
        self.showStudent()
        self.__oveallGPA = (int)( (self.__math + self.__english + self.__science + self.__social) / 400 * 100 ); 
        print(self.__name,self.__rollno, self.__overallGPA)    
        
def main():
    print("Student: ")
    s = Student()
    s.setStudent()
    s.showStudent()        
    s.printResult()
    
    
if __name__=="__main__":
    main() 
    
```
