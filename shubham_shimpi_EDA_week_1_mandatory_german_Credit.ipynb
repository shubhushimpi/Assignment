## Q4.How many highly skilled men (below 35 years) are staying in their own house?

```python
job=pd.get_dummies(credit_eda_df["Job"])

job=job.drop(["skilled","unskilled and non-resident","unskilled and resident"],axis=1)

job=job.replace({1:True,0:False})

age=credit_eda_df[['Age']]<35

housing=pd.get_dummies(credit_eda_df[["Housing"]])

Housing_own=housing.drop(["Housing_free","Housing_rent"],axis=1)

Housing_own=Housing_own.replace({1:True,0:False})

skilled_housing_age=pd.concat([Housing_own,age,job],axis=1)

X=(skilled_housing_age["Housing_own"]) & (skilled_housing_age["Age"] & (skilled_housing_age["highly skilled"]))

X=X.replace({True:"Yes", False:"No"})

X=pd.get_dummies(X)

X=X.drop("No",axis=1)

y=X["Yes"].tolist()

print("Highly skilled men (below 35 years) are staying in their own house:",y.count(1))
```

So Their are 44 highly skilled  men below 35 years are staying in their own house.
